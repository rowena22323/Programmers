```r
library(tidytext)
library(textdata)
library(tidyverse)
library(tm)
library(stringr)
library(dplyr)
library(textstem)
```

```r
txt1<-readLines("austene/946.txt")
txt1 <- lemmatize_words(txt1)
txt1 <- as.character(txt1)
```

```r
my.df.text<-data_frame(paper.id = 1, doc=txt1)
my.df.text.word <- my.df.text %>% unnest_tokens(word, doc)
my.result1 <- my.df.text.word %>% inner_join(get_sentiments("afinn")) %>% count(word, paper.id, value)
```

```r
ladysusan<-my.result1$value
```
